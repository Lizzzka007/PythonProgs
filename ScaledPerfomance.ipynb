# Div2x

```python
def Scale1(arr, ref):
    n = len(arr)
    for i in range(n):
        arr[i] /= ref[i]

def Scale(arr):
    n = len(arr)
    for i in range(n):
        arr[i] = math.log10(arr[i])
```
