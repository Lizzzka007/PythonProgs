```python
import matplotlib.pyplot as plt
import math
```

# Div2x

```python
def Scale1(arr, ref):
    n = len(arr)
    for i in range(n):
        arr[i] /= ref[i]

def Scale(arr):
    n = len(arr)
    for i in range(n):
        arr[i] = math.log10(arr[i])
```

```python
GPU_d =     [0.002621, 0.006200, 0.031509, 0.057759, 0.391489, 3.034166, 24.529760]
CPU_n1_d  = [0.00464, 0.03722, 0.27189, 0.52578, 4.57393, 57.45157, 449.75486]
CPU_n14_d = [0.00049, 0.00335, 0.02437, 0.03879, 0.40486, 16.25225, 138.76755]
AngleStep = [4, 2, 1, 0.8, 0.4, 0.2, 0.1] 

# Scale(GPU_d, CPU_n1_d)
# Scale(CPU_n14_d, CPU_n1_d)
# Scale(GPU, CPU_n1)

Scale(GPU_d)
Scale(CPU_n14_d)
Scale(CPU_n1_d)

offset = abs(min(CPU_n14_d))


fig, ax = plt.subplots(figsize=(12,12))
plt.xlabel("Шаг сетки, градусы", fontsize=24)
plt.ylabel("$\dfrac{time}{CPU \;\; n1 \;\; time}$", fontsize=24)
# ax.plot(AngleStep, CPU_n1, marker='o', label="CPU n1")
ax.plot(AngleStep, [offset + i for i in CPU_n14_d], marker='o', linewidth=3, label="CPU n14")
ax.plot(AngleStep, [offset + i for i in GPU_d], marker='o', linewidth=3, label="GPU")
ax.plot(AngleStep, [offset + i for i in CPU_n1_d], marker='o', linewidth=3, label="CPU n1")
ax.legend(fontsize = 24, loc='best')
plt.title("Div-x2", fontsize=36);
```
