```python
def Scale(arr, ref):
    n = len(arr)
    for i in range(n):
        arr[i] /= ref[i]
```

```python
GPU =     [0.002621, 0.006200, 0.031509, 0.057759, 0.391489, 3.034166, 24.529760]
CPU_n1  = [0.00464, 0.03722, 0.27189, 0.52578, 4.57393, 57.45157, 449.75486]
CPU_n14 = [0.00049, 0.00335, 0.02437, 0.03879, 0.40486, 16.25225, 138.76755]
AngleStep = [4, 2, 1, 0.8, 0.4, 0.2, 0.1] 

Scale(CPU_n1, GPU)
Scale(CPU_n14, GPU)
Scale(GPU, GPU)

fig, ax = plt.subplots(figsize=(12,12))
plt.xlabel("Шаг сетки, градусы", fontsize=16)
plt.ylabel("Время, секунды", fontsize=16)
ax.plot(AngleStep, CPU_n1, marker='o', label="CPU n1")
ax.plot(AngleStep, CPU_n14, marker='o', label="CPU n14")
ax.plot(AngleStep, GPU, marker='o', label="GPU")
ax.legend(fontsize = 16, loc='best')
plt.title("div-x2", fontsize=24);
```
